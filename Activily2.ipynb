## MODULES

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
```

## FUNCTION MEAN SQUARED

```python

def mean_squared_displacement(traj_df):

    # Posici√≥n inicial
    initial_position = traj_df[['x_pos', 'y_pos']].iloc[0]
    
    # Calcular el desplazamiento al cuadrado para cada instante de tiempo
    displacement_squared = np.sum((traj_df[['x_pos', 'y_pos']] - initial_position) ** 2, axis=1)
    
    return displacement_squared

```

## READ FILES

```python
BM_traj = pd.read_csv('brownian_3.csv')  
CRW_traj = pd.read_csv('crw_6_9.csv')    
```

## CALCULATE MSD

```python
BM_msd = mean_squared_displacement(BM_traj)
CRW_msd = mean_squared_displacement(CRW_traj)
```

## PLOT

```python
fig_msd = go.Figure()

fig_msd.add_trace(go.Scatter( x = BM_traj.index, y = BM_msd, mode = 'lines', name = 'BM - Mean Squared Displacement'))

fig_msd.add_trace(go.Scatter( x = CRW_traj.index, y = CRW_msd, mode = 'lines',name = 'CRW - Mean Squared Displacement'))

fig_msd.update_layout( title = 'Mean Squared Displacement: BM vs CRW', xaxis_title = 'Time Step', yaxis_title = 'MSD',showlegend = True)

fig_msd.show()
```
